```python
# return Nan if index is not found
def ensured_access(values, idx):
    result = values
    try:
        for i in idx:
            result = result[i]
        return result
    except IndexError or KeyError:
        return pd.np.nan

movies_tmp = movies.copy()
movies_tmp['budget'] = movies['budget']/1000000
movies_tmp['profit'] = (movies['revenue']-movies['budget'])/1000000
movies_tmp['year'] = pd.to_datetime(movies_tmp['release_date']).apply(lambda x: x.date().year)
movies_tmp['month'] = pd.to_datetime(movies_tmp['release_date']).apply(lambda x: x.date().month)
# The first production country is assumed
movies_tmp['country'] = movies_tmp['production_countries'].apply(lambda x: ensured_access(x, [0, 'name']))
movies_tmp['language'] = movies_tmp['spoken_languages'].apply(lambda x: ensured_access(x, [0, 'name']))
movies_tmp['Director'] = credits['crew'].apply(lambda y: ensured_access([x['name'] for x in y if x['job'] == 'Director'], [0]))
movies_tmp['Actor1'] = credits['cast'].apply(lambda x: ensured_access(x, [1, 'name']))
movies_tmp['Actor2'] = credits['cast'].apply(lambda x: ensured_access(x, [2, 'name']))
movies_tmp['Actor3'] = credits['cast'].apply(lambda x: ensured_access(x, [3, 'name']))
movies_tmp['genres'] = movies_tmp['genres'].apply(lambda genre: '|'.join([x['name'] for x in genre]))
movies_tmp['keywords'] = movies_tmp['keywords'].apply(lambda keyword: '|'.join([x['name'] for x in keyword]))
```

```python
#https://www.kaggle.com/ibtesama/getting-started-with-a-movie-recommendation-system
df_original = df
C= df_original['vote_average'].mean()
m= df_original['vote_count'].quantile(0.9)
```

```python
def weighted_rating(x, m=m, C=C):
    v = x['vote_count']
    R = x['vote_average']
    # Calculation based on the IMDB formula
    return (v/(v+m) * R) + (m/(m+v) * C)
```

```python
#this will get movies in sorted order by score
def getMoviesByScore(df_original):
    df_byScore = df_original.copy().loc[df_original['vote_count'] >= m]
    df_byScore['score'] = df_byScore.apply(weighted_rating, axis=1)
    df_byScore = df_byScore.sort_values('score', ascending=False)
    return df_byScore[:10]
```

```python
def findMaxNum(x):
    max=x[0]
    for i in x:
        if(i > max):
            max=i
        
    return max
```

```python
def convertReleaseDateToInt(x):
    
    b=str(x)
    
    c=b[:4]
    d=pd.to_numeric(c,errors='coerce')
    
    return d
```

```python
#trendingByYear Stores the popular movies of the recent year 

def getMoviesByYear(df_original,year):
    df_byYear=df_original
    df_byYear.dropna(subset=['release_date'],axis=0)
    df_byYear['year']=df_byYear['release_date'].apply(convertReleaseDateToInt)
    max=year
    df_byYear = df_byYear.copy().loc[df_byYear['year'] == max]
    df_byYear = df_byYear.sort_values('popularity', ascending=False)
    return df_byYear[:10]
df_byYear=getMoviesByYear(df_original,2015)
df_byYear[['title', 'vote_count', 'vote_average', 'popularity','year']].head(10)
```

```python
def convertGenresToList(x):
    list=[]
    a=str(x)
    b=a.split('"}')
    for i in b:
        c=i.split('"name": "')
        if(len(c)>1):
            list.append(c[1])
    return list

def check(x, name):
    
    chk = False
    for i in x:
            if(i==name):
                chk=True
    return chk
```

<h5> Function to sort movies by each Genre</h5>
This function will return a dataset of movies in sorted order for certain genre by popularity.

```python
def getMoviesByGenre(x,name):
    
    x['GenresName']=x['genres'].apply(convertGenresToList)
    x['GenresTrue']=[check(i,name) for i in x['GenresName']]
    
    x = x.copy().loc[x['GenresTrue'] == True]
    
    x= x.sort_values('popularity', ascending=False)
    return x[:20]
```

```python
# adding a new column 'text' to df 
df['text'] = df['keywords']


# Jasser Code:  PARSING KEYWORDRS  +  'text' COLUMN CREATION (all relative text to the movie)
#                                      keywords + overview + tagline + title 
import json

for i in df.index:
    data = json.loads(df.keywords[i])
    key_str = ""
    for x in data:
        key_str += " "+x['name']
    key_str+= " "+df.overview[i]
    key_str+= " "+df.tagline[i]
    key_str+= " "+df.title[i]
    df.text[i] = key_str



   
 
# Jasser Code:  REMOVING STOP_WORDS + PUNCTUATIONS

# Install relative packages to nltk
# import nltk
# nltk.download()

from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize 
import string


# LOADING STOP_WORDS from nltk library 
stop_words = set(stopwords.words('english')) 

for i in df.index:
    word_tokens = word_tokenize(df.text[i]) 
    filtered_sentence = [w for w in word_tokens if not w in stop_words and w not in string.punctuation] 
    concat_str = ""
    for word in filtered_sentence:
        concat_str+= " "+word
    df.text[i]= concat_str


# TF-IDF 
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()

response = tfidf.fit_transform(df.text.tolist())

# --- Uncomment for testing 
#feature_names = tfidf.get_feature_names()
#for col in response.nonzero()[1]:
#    print (feature_names[col], ' - ', response[0, col])

# let's drop some column for clarity 
df = df.drop(['keywords','overview', 'tagline'], axis=1)

#---------------------------------------------------------------
# Jasser Code: PROCESSING GENRE 

genres = []
for i in df.index:
    data = json.loads(df.genres[i])
    for x in data:
        if x['name'] not in genres:
            genres.append(x['name'])

# print(genres)

for genre in genres:
    df[genre] = df['vote_count']
    df[genre].values[:] = 0
    
for i in df.index:
    data = json.loads(df.genres[i])
    for x in data:
        df[x['name']][i] = 1


df_Genre = df[genres].copy()
df = df.drop(genres, axis=1)
#print(df_Genre.head())



#-----------------------------------------------------------------
# Jasser Code: PROCESSING ACTORS (top 3 actors for every movie)

topActors = []
for i in df.index:
    data = json.loads(df.cast[i])
    count= 0
    for x in data:
        count+=1 
        if x['name'] not in topActors:
            topActors.append(x['name'])
        if count == 3:
            break

for actor in topActors:
    df[actor] = df['vote_count']
    df[actor].values[:] = 0
    
for i in df.index:
    data = json.loads(df.cast[i])
    count= 0
    for x in data:
        count +=1
        df[x['name']][i] = 1
        if count == 3:
            break

df_Actors = df[topActors].copy()
df = df.drop(topActors, axis=1)
#print(df_Actors.head(50))

#--------------------------------------------------------------
# Jasser Code : 
# Other Factors that we can consider 

df_Other = df[['revenue','popularity','runtime','release_date']].copy()
df_Other['release_date']= pd.to_datetime(df_Other['release_date'])
df_Other['YearMonth'] = df_Other['release_date'].map(lambda x: 100*x.year + x.month)
df_Other = df_Other.drop(['release_date'], axis=1)
#print(df_Other.head())

from sklearn import preprocessing

x = df_Other.values
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
df_OtherScaled = pd.DataFrame(x_scaled,columns=df_Other.columns)
#print(df_OtherScaled)
print("Content Based Processing is FINISHED")
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def getRecommendations(Movie, df, response, df_Genre, df_Actors, df_OtherScaled):
    selectedMovie = Movie
    indexSelectedMovie = df.index[df['title'] == selectedMovie][0]
    
    # Text Similarity
    MovieTextSim = []
    for i in range(0,len(df.index)):
        cos_Sim1 = cosine_similarity(response[indexSelectedMovie], response[i])[0][0]
        MovieTextSim.append(cos_Sim1)
    
    # Genre similarity
    vec1 = np.array([df_Genre.loc[indexSelectedMovie,:].tolist()])
    MovieGenreSim = []
    for i in range(0,len(df.index)):
        vec2 = np.array([df_Genre.loc[i,:].tolist()])
        cos_Sim2 = cosine_similarity(vec1, vec2)[0][0]
        MovieGenreSim.append(cos_Sim2)
        
    # Actor similarity 
    vec1 = np.array([df_Actors.loc[indexSelectedMovie,:].tolist()])
    MovieActorSim = []
    for i in range(0,len(df.index)):
        vec2 = np.array([df_Actors.loc[i,:].tolist()])
        cos_Sim3 = cosine_similarity(vec1, vec2)[0][0]
        MovieActorSim.append(cos_Sim3)
        
    # Other Factor Similarity
    vec1 = np.array([df_OtherScaled.loc[indexSelectedMovie,:].tolist()])
    MovieOtherSim = []
    for i in range(0,len(df.index)):
        vec2 = np.array([df_OtherScaled.loc[i,:].tolist()])
        cos_Sim4 = cosine_similarity(vec1, vec2)[0][0]
        MovieOtherSim.append(cos_Sim4)
        
    # Calculating Average Similarity
    AveSim = []
    for i in range(0,len(MovieOtherSim)):
        AVG = 0.5 * MovieTextSim[i] + 0.25 * MovieGenreSim[i] + 0.15 * MovieActorSim[i] + 0.10 * MovieOtherSim[i]
        AveSim.append(AVG)
    arr = np.array(AveSim)
    # top 10 Movies
    indices = arr.argsort()[-11:][::-1] 
    indices = indices[1:]
    RecommendedMovies = []
    for idx in indices:
        RecommendedMovies.append(df.loc[idx,'title'])
    
    return RecommendedMovies
        

    
```

```python
def CF_getRecommendations(userID, ratings, svd):
    movieIDs = ratings.movieId.unique().tolist()
    df_tmp = ratings[ratings['userId'] == userID]
    ratedList = df_tmp.movieId.tolist()

    nonRatedList = [x for x in movieIDs if x not in ratedList]

    estimatedRatings = []
    for nonRatedMovieID in nonRatedList:
        estimatedRatings.append(svd.predict(userID, nonRatedMovieID).est)

    arr = np.array(estimatedRatings)
    # top 10 Movies
    indices = arr.argsort()[-10:][::-1]
    MoviesIDsToRecommend = []
    for x in indices:
        MoviesIDsToRecommend.append(nonRatedList[x])
        
    return MoviesIDsToRecommend
```
